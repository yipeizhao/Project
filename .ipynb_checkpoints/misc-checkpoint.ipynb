```python
import utilities as ut
import networkx as nx
import Complexity as cx
import matplotlib.pyplot as plt
```

```python
graphs,df = ut.random_networks(15,False,100)
```

```python
c1est = [cx.C1est(g) for g in graphs]
c1espec = [cx.C1espec(g) for g in graphs]
c2espec = [cx.C2espec(g) for g in graphs]
```

```python
fig,axes = plt.subplots(1, 3,figsize = ([15,5]))
axes[0].scatter(df["Number_of_edges"],c1est,s=10,color = "black")
axes[1].scatter(df["Number_of_edges"],c1espec,s=10,color = "black")
axes[2].scatter(df["Number_of_edges"],c2espec,s=10,color = "black")
for item in axes:
    item.set_yticks([0,0.5,1,1.5])
    item.set_xlabel("m")
axes[0].set_ylabel("C1e,st")
axes[1].set_ylabel("C1e,spec")
axes[2].set_ylabel("C2e,spec")
```

```python
n = [6,7,8]
c2graphs= [];dfs= []
for item in n:
    g,d = ut.random_networks(item,True,50)
    c2graphs.append(g);dfs.append(d)
```

```python
c2especs = []
for item in c2graphs:
    temp_result = [cx.C2espec(g) for g in item]
    c2especs.append(temp_result)
```

```python
fig,axes = plt.subplots(1,3,figsize = (15,5))
for i in range(len(n)):
    axes[i].scatter(dfs[i]["Number_of_edges"],c2especs[i],s=10,color = "black")
    axes[i].set_yticks([0,0.5,1])
    axes[i].set_title("n="+str(n[i]))
    axes[i].set_xlabel("m")
axes[0].set_ylabel("$C_{1e,st}$")
axes[1].set_ylabel("$C_{1e,spec}$")
axes[2].set_ylabel("$C_{2e,spec}$")
```

```python
n = [7,8,9,10]
graphs_list = [];dfs_list = []
for item in n:
    temp_g,temp_df = ut.random_networks(item,True,50)
    graphs_list.append(temp_g)
    dfs_list.append(temp_df)
```

```python
results = []
for i in range(len(n)):
    result = [cx.MAg(g) for g in graphs_list[i]]
    results.append(result)
```

```python
fig,axes = plt.subplots(1,4,figsize = (20,5))
for i in range(len(n)):
    axes[i].scatter(dfs_list[i]["Number_of_edges"],results[i],s=10,color="blue",alpha = 0.5,label = "$MA_g$",marker= "x")
    axes[i].set_xlabel("$m$")
    axes[i].set_ylabel("$MA_g$")
    axes[i].set_yticks([0,0.5,1])
    axes[i].set_title("n="+str(n[i]))
```

```python
results_mari = []
for i in range(len(n)):
    result = [cx.MAri(g) for g in graphs_list[i]]
    results_mari.append(result)
```

```python
fig,axes = plt.subplots(1,4,figsize = (20,5))
for i in range(len(n)):
    axes[i].scatter(dfs_list[i]["Number_of_edges"],results[i],s=10,color="blue",alpha = 0.5,label = "$MA_g$",marker= "x")
    axes[i].scatter(dfs_list[i]["Number_of_edges"],results_mari[i],s=10,color="red",label="$MA_{ri}$")
    axes[i].set_xlabel("$m$")
    axes[i].set_ylabel("$Complexity$")
    axes[i].set_yticks([0,0.5,1])
    axes[i].set_title("n="+str(n[i]))
axes[0].legend()
axes[3].legend()
```
