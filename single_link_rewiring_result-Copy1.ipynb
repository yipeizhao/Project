```python
import pandas as pd
import utilities as ut
import Complexity as cx
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load networks
load_path = "real_networks/processed/"
index = ["dolphins","pdzbase","GBPT_train",
         "hamsterster","Roget","flight"]
load_path_bus = "real_networks/processed/modified_bus/m_"
index_bus = ["london","paris","berlin","sydney","detroit","beijing"]
dfs = [pd.read_csv(load_path+item+".csv") for item in index]
dfs_bus = [pd.read_csv(load_path_bus+item+".csv") for item in index_bus]
networks = [ut.df_to_network(item) for item in dfs]
networks_bus = [ut.df_to_network(item) for item in dfs_bus]
```

```python
#Load networks
load_path = "real_networks/processed/"
index = ["dolphins","pdzbase","GBPT_train",
         "hamsterster","Roget","flight"]
load_path_bus = "real_networks/processed/modified_bus/m_"
index_bus = ["london","paris","berlin","sydney","detroit","beijing"]
```

```python
load_path = "result/single_edge/"
results = [pd.read_csv(load_path + item + ".csv") for item in index]
results = [item["result"] for item in results]

results_bus = [pd.read_csv(load_path + item + ".csv") for item in index_bus]
results_bus = [item["result"] for item in results_bus]
```

```python
load_path = "result/pairwise/"
results1 = [pd.read_csv(load_path + item + ".csv") for item in index]
results1 = [item["result"] for item in results1]

results_bus1 = [pd.read_csv(load_path + item + ".csv") for item in index_bus]
results_bus1 = [item["result"] for item in results_bus1]
```

```python
step_prob = 0.05
prob_list = np.linspace(0,1,int(1/step_prob))
sample = 10
```

```python
fig,ax = plt.subplots(6,2,figsize = [10,24])
fig.suptitle("Single edge rewiring")
for i in range(len(networks)):
    ax[i][0].plot(prob_list,results[i])
    ax[i][1].plot(prob_list,results_bus[i])
for i in range(2):
    for j in range(6):
        ax[j][i].set_xticks([])
        #ax[j][i].set_yticks([])
ax[0][0].set_title("Real networks")
ax[0][1].set_title("Bus networks(modified)")
for i in range(6):
    ax[i][0].set_ylabel("OdC Complexity")
for i in range(2):
    ax[5][i].set_xlabel("Proabability of rewiring")

#fig,ax = plt.subplots(6,2,figsize = [10,24])
#fig.suptitle("Single edge rewiring")
for i in range(len(networks)):
    ax[i][0].plot(prob_list,results1[i],color = "red")
    ax[i][1].plot(prob_list,results_bus1[i],color = "red")
for i in range(2):
    for j in range(6):
        ax[j][i].set_xticks([])
        #ax[j][i].set_yticks([])
ax[0][0].set_title("Real networks")
ax[0][1].set_title("Bus networks(modified)")
#for i in range(6):
#    ax[i][0].set_ylabel("OdC Complexity")
#for i in range(2):
#    ax[5][i].set_xlabel("Proabability of rewiring")
```

Codea below generatethe dataset

```python
#Dolphins:
dol_result = [0]*len(prob_list)
for iter_n in range(sample):
    G1 = networks[0].copy()
    for i in range(int(1/step_prob)):
        G2 = G1.copy()
        G1 = ut.single_link_rewiring(G2, step_prob)
        dol_result[i] = dol_result[i] + cx.OdC(G1)
dol_result = [item/sample for item in dol_result]
df = pd.DataFrame(data = {"result" : dol_result})
df.to_csv("result/single_edge/"+str(index[0])+".csv")
```
