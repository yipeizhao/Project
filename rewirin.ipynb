```python
from rewiring import rewiring
import matplotlib.pyplot as plt
```

```python
index = ["dolphins","pdzbase","GBPT_train",
         "hamsterster","Roget","flight"]
index_bus = ["london","paris","berlin","sydney","detroit","beijing"]
```

```python
real_networks = [rewiring(item) for item in index]
bus_networks = [rewiring(item,bus = True) for item in index_bus]
```

```python
index = ["dolphins","pdzbase","GBPT_train",
         "hamsterster","Roget"]
temp_networks = [rewiring(item) for item in index]
```

```python
temp_networks[1].experiment("S",30)
temp_networks[1].write_result("S")
```

```python
temp_networks[0].experiment("S",75)
temp_networks[0].write_result("S")
temp_networks[0].experiment("P",75)
temp_networks[0].write_result("P")
```

```python
for item in temp_networks:
    item.experiment("P",15)
    item.experiment("S",15)
    item.write_result("P")
    item.write_result("S")
```

```python
for item in real_networks:
    plt.figure()
    item.plot_result("S+P")
```

```python
for item in bus_networks:
    plt.figure()
    item.plot_result("S+P")
```
